````python
# write a program to check given number is even
num=int(input("please enter a number"))
if (num%2 == 0):
    print(num,"is even number")
else:
    print(num,"is odd number")
    
````

````python
# write a program to find the given number is positive or negative or zero?
num=int(input("please enter a number"))
if (num == 0):
    print(num,"is zero ")
elif (num > 0):
    print(num,"is positive")
else:
    print (num," is negative")
    
````

````python
"""
write a program to find the grades based on student marks
1. if marks in between 80 to 100 grade A
2.if marks in between 60 to 80 grade B
3. if marks in between 40 to 60 grade C
4. if marks in between 25 to 40 grade D
5. if marks in between 0 to 25 grade Fail
6. if marks other than above output should be please check valid marks
"""
marks=int(input("please enter marks"))
if (marks>=80 and marks<=100):
    print ("Grade A")
elif (marks>=60 and marks<=80):
    print ("Grade B") 
elif (marks>=40 and marks<=60):
    print ("Grade C")    
elif (marks>=24 and marks<=40):
    print ("Grade D")
elif (marks>=0 and marks<=25):
    print ("Grade Fail")
else :
      print("please check the valid marks")
      
````

````python
n1=int(input("please enter 1st number:"))
n2=int(input("please enter 2st number:"))
n3=int(input("please enter 3st number:"))
if (n1>=n2 and n1>=n3):
    print(n1,"is largest the number")
elif(n2>=n1 and n2>=n3):
    print(n2,"is largest the number")
else:
    print(n3,"is largest the number")
    
````
