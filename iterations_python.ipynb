````python
#write a program to iterate 1 to 10 numbers
for i in range(1,10):
    print(i)
````

````python
# write a program to print names using for loop
names=["shiva","annie","vikas","ram"]
for n in names:
    print("my name is",n)
    
````

````python
#write a program to find the vowels from string using for loop.
name=str(input("please enter your name here:"))
print(name)
for char in name :
    if (char in ["a",'e','i','o','u']):
        print(char)
````

````python
#write a program to find the non vowels from string using for loop
name=str(input("please enter your name here:"))
print(name)
for char in name :
    if (char not in ["a",'e','i','o','u']):
        print(char,end=",")
````

````python
# write a program to find the sum of all numbers in between 1 to 20
# Initialize sum variable
total = 0

# Loop through numbers 1 to 20 (inclusive)
for number in range(1, 21):
    total += number

# Print the result
print("The sum of numbers from 1 to 20 is:", total)
````

````python
# write a program to find the sum of all odd numbers in between 1 to 1000 using for loop
sum_odds = 0
for num in range(1, 1001):
    if num % 2 != 0:
        sum_odds += num
print("Sum of all odd numbers from 1 to 1000:", sum_odds)
````

````python
# write a program to find the product of all even numbers in between 500 to 600 using for loop.
product_even = 1
for num in range(500, 601):
    if num % 2 == 0:
        product_even *= num
print("product of all even numbers from 1 to 1000:", product_even)
````

````python
# write a program to find the numbers in between 1 to 20 using while loop
sn=1
en=20
while (sn<=en):
    print(sn,end=",")
    sn=sn+1
````

````python
#write a program to print odd numbers odd numbers in between 100 to 200 using while loop.
num = 100
while num <= 200:
    if num % 2 != 0:
        print(num,end=",")
    num += 1
````

````python
# write a program to find the sum of all even numbers in between 1 to 100 using while loop
num = 1
sum_evens = 0
while num <= 100:
    if num % 2 == 0:
        sum_evens += num
    num += 1
print("Sum of all even numbers from 1 to 100:", sum_evens)
````

````python
# write a program to find the product of all odd numbers in between 10 to 30 using while loop
num = 10
product = 1
while num <= 30:
    if num % 2 != 0:
        product *= num
    num += 1
print("Product of all odd numbers from 10 to 30:", product)
````

````python
#write a program to find the sum of odd numbers and product of all even numbers from 1 to 100
num = 1
sum_odds = 0
product_evens = 1

while num <= 100:
    if num % 2 != 0:
        sum_odds += num
    else:
        product_evens *= num
    num += 1

print("Sum of all odd numbers from 1 to 100:", sum_odds)
print("Product of all even numbers from 1 to 100:", product_evens)
````
